# Probability Distributions:


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Normal Distribution (Gaussian) :

```python

# Normal Distribution (Gaussian)
mu = 0
sigma = 1
x = np.linspace(-5, 5, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), label='Normal')
plt.title('Normal Distribution')
plt.legend()
plt.show()
```

# Uniform Distribution:

```python
# Uniform Distribution
a = 0
b = 10
x = np.linspace(-2, 12, 100)
plt.plot(x, stats.uniform.pdf(x, a, b - a), label='Uniform')
plt.title('Uniform Distribution')
plt.legend()
plt.show()
```

# Exponential Distribution:

```python
# Exponential Distribution
scale = 1
x = np.linspace(0, 10, 100)
plt.plot(x, stats.expon.pdf(x, scale=scale), label='Exponential')
plt.title('Exponential Distribution')
plt.legend()
plt.show()
```

# Poisson Distribution :

```python
# Poisson Distribution
mu = 3
x = np.arange(0, 10)
plt.stem(x, stats.poisson.pmf(x, mu), basefmt='C0-', label='Poisson')
plt.title('Poisson Distribution')
plt.legend()
plt.show()
```

# Binomial Distribution:

```python
# Binomial Distribution
n = 10
p = 0.3
x = np.arange(0, n+1)
plt.stem(x, stats.binom.pmf(x, n, p), basefmt='C0-', label='Binomial')
plt.title('Binomial Distribution')
plt.legend()
plt.show()
```

# Beta Distribution:

```python
# Beta Distribution
alpha = 2
beta = 5
x = np.linspace(0, 1, 100)
plt.plot(x, stats.beta.pdf(x, alpha, beta), label='Beta')
plt.title('Beta Distribution')
plt.legend()
plt.show()
```

# Gamma Distribution:

```python
# Gamma Distribution
alpha = 2
beta = 1
x = np.linspace(0, 10, 100)
plt.plot(x, stats.gamma.pdf(x, alpha, scale=1/beta), label='Gamma')
plt.title('Gamma Distribution')
plt.legend()
plt.show()
```

# Chi-Square Distribution

```python
# Chi-Square Distribution
df = 3
x = np.linspace(0, 10, 100)
plt.plot(x, stats.chi2.pdf(x, df), label='Chi-Square')
plt.title('Chi-Square Distribution')
plt.legend()
plt.show()
```

# Student's t-Distribution:

```python
# Student's t-Distribution
df = 5
x = np.linspace(-5, 5, 100)
plt.plot(x, stats.t.pdf(x, df), label="t-Distribution")
plt.title("Student's t-Distribution")
plt.legend()
plt.show()
```

# Geometric Distribution:

```python
# Geometric Distribution
p = 0.3
x = np.arange(1, 11)
plt.stem(x, stats.geom.pmf(x, p), basefmt='C0-', label='Geometric')
plt.title('Geometric Distribution')
plt.legend()
plt.show()
```

# Hypergeometric Distribution

```python
# Hypergeometric Distribution
N = 20
K = 8
n = 5
x = np.arange(max(0, n - (N - K)), min(n, K) + 1)
plt.stem(x, stats.hypergeom.pmf(x, N, K, n), basefmt='C0-', label='Hypergeometric')
plt.title('Hypergeometric Distribution')
plt.legend()
plt.show()
```

# Log-Normal Distribution:

```python
# Log-Normal Distribution
mu = 0
sigma = 1
x = np.linspace(0.01, 10, 100)
plt.plot(x, stats.lognorm.pdf(x, sigma, scale=np.exp(mu)), label='Log-Normal')
plt.title('Log-Normal Distribution')
plt.legend()
plt.show()

```

# Probability Mass Function (Discrete Distribution)



```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import poisson

lambda_ = 3  # Mean of the Poisson distribution
x = np.arange(0, 15)
pmf = poisson.pmf(x, lambda_)

sns.barplot(x=x, y=pmf)
plt.title("Poisson Distribution (PMF) with Seaborn")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```
