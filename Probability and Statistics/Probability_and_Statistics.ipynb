# Calculating Mean and Standard Deviation


```python
import numpy as np

data = np.array([15, 20, 25, 30, 35, 40, 45])

mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)
```

# Hypothesis Testing (t-test)

```python
import numpy as np
from scipy.stats import ttest_ind

group1 = np.array([23, 25, 28, 32, 35])
group2 = np.array([18, 20, 22, 28, 30])

t_statistic, p_value = ttest_ind(group1, group2)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

# Hypothesis Testing (Chi-Square Test)


```python
# Hypothesis Testing (Chi-Square Test)
import numpy as np
from scipy.stats import chi2_contingency

# Create a contingency table for observed and expected values
observed = np.array([[30, 45],
                     [15, 10]])

chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-Square:", chi2)
print("P-value:", p)
```

# Correlation and Scatter

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate correlated data
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)

# Calculate Pearson correlation coefficient
correlation_coefficient = np.corrcoef(x, y)[0, 1]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

print("Correlation Coefficient:", correlation_coefficient)
```

# Confidence Intervals

```python
import numpy as np
from scipy.stats import norm

def confidence_interval_func(confidence_level, sample_mean, sample_std, data):
  z_critical = norm.ppf(1 - (1 - confidence_level) / 2)

  margin_of_error = z_critical * (sample_std / np.sqrt(len(data)))
  confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
  return z_critical, margin_of_error, confidence_interval
data = np.array([28, 32, 35, 38, 40, 42, 45, 48, 50, 52])

confidence_level= 0.90
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
z_critical, margin_of_error, confidence_interval = confidence_interval_func(confidence_level, sample_mean, sample_std, data)
print(f"Confidence Level: {confidence_level} , Confidence Interval: {confidence_interval}")
confidence_level= 0.95
z_critical, margin_of_error, confidence_interval = confidence_interval_func(confidence_level, sample_mean, sample_std, data)
print(f"Confidence Level: {confidence_level} , Confidence Interval: {confidence_interval}")
confidence_level= 0.99
z_critical, margin_of_error, confidence_interval = confidence_interval_func(confidence_level, sample_mean, sample_std, data)
print(f"Confidence Level: {confidence_level} , Confidence Interval: {confidence_interval}")
```

# Probability Density Estimation (Kernel Density Estimation)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity

# Generate data from a normal distribution
np.random.seed(0)
data = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(5, 1, 100)])

# Fit a kernel density estimator
kde = KernelDensity(kernel='gaussian', bandwidth=0.5).fit(data[:, None])

x_vals = np.linspace(-3, 8, 1000)
log_density = kde.score_samples(x_vals[:, None])

plt.plot(x_vals, np.exp(log_density))
plt.title("Kernel Density Estimation")
plt.xlabel("X")
plt.ylabel("Probability Density")
plt.show()
```

# Central Limit Theorem Illustration

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data from a non-normal distribution
np.random.seed(0)
data = np.random.exponential(2, 1000)

sample_means = []
num_samples = 1000
sample_size = 30

for _ in range(num_samples):
    sample = np.random.choice(data, size=sample_size)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=20)
plt.title("Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

# Analysis of Variance (ANOVA)

```python
import numpy as np
from scipy.stats import f_oneway

# Simulated data for three groups
group1 = np.random.normal(50, 10, 50)
group2 = np.random.normal(55, 10, 50)
group3 = np.random.normal(60, 10, 50)

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_statistic)
print("P-value:", p_value)
```
